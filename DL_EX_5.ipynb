```python
print("Reg No: URK23AI006")
# time_series_rnn_cleaned.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Load dataset
data = pd.read_csv("/content/daily-minimum-temperatures-in-me.csv")
data.columns = data.columns.str.strip()
data['Date'] = pd.to_datetime(data['Date'])

# Clean numeric column (remove ? and convert to float)
data['Daily minimum temperatures'] = (
    data['Daily minimum temperatures']
    .astype(str)
    .str.replace('?', '', regex=False)
    .astype(float)
)

# Extract temperature values
values = data['Daily minimum temperatures'].values.reshape(-1, 1)

# Normalize data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)

# Create sequences
X, y = [], []
time_step = 10
for i in range(len(scaled) - time_step):
    X.append(scaled[i:i + time_step])
    y.append(scaled[i + time_step])
X, y = np.array(X), np.array(y)

# Build RNN model
model = Sequential([
    SimpleRNN(64, activation='relu', input_shape=(time_step, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=20, batch_size=16, verbose=1)

# Predict and inverse transform
predicted = model.predict(X)
predicted = scaler.inverse_transform(predicted)
real = scaler.inverse_transform(y)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(real, label='Actual', color='blue')
plt.plot(predicted, label='Predicted', color='red')
plt.legend()
plt.title('Daily Minimum Temperature Prediction using RNN')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
print("Reg No: URK23AI006")
# car_sales_rnn_corrected.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Load dataset
data = pd.read_csv("/content/Car_sales.csv")
data.columns = data.columns.str.strip()  # clean any stray spaces

print("Columns available:", data.columns.tolist())

# Use correct column for sales
sales = data['Sales_in_thousands'].astype(float).values.reshape(-1, 1)

# Normalize
scaler = MinMaxScaler(feature_range=(0, 1))
sales_scaled = scaler.fit_transform(sales)

# Create sequences (12-month lookback)
X, y = [], []
time_step = 12
for i in range(len(sales_scaled) - time_step):
    X.append(sales_scaled[i:i + time_step])
    y.append(sales_scaled[i + time_step])
X, y = np.array(X), np.array(y)

# Build RNN model
model = Sequential([
    SimpleRNN(100, activation='tanh', input_shape=(time_step, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=30, batch_size=8, verbose=1)

# Predictions
predicted = model.predict(X)
predicted = scaler.inverse_transform(predicted)
actual = scaler.inverse_transform(y)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(actual, label='Actual Sales', color='blue')
plt.plot(predicted, label='Predicted Sales', linestyle='--', color='red')
plt.title('Car Sales Prediction using RNN')
plt.xlabel('Time Step')
plt.ylabel('Sales (in thousands)')
plt.legend()
plt.show()
```

```python
print("Reg No: URK23AI006")
# stock_price_rnn.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

data = pd.read_csv("/content/stock_prices.csv")
data.columns = data.columns.str.strip().str.lower()
data['date'] = pd.to_datetime(data['date'])
data = data.sort_values('date')

prices = data['close'].values.reshape(-1, 1)

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_prices = scaler.fit_transform(prices)

time_step = 30
X, y = [], []
for i in range(time_step, len(scaled_prices)):
    X.append(scaled_prices[i-time_step:i, 0])
    y.append(scaled_prices[i, 0])
X, y = np.array(X), np.array(y)
X = X.reshape(X.shape[0], X.shape[1], 1)

model = Sequential([
    SimpleRNN(80, activation='relu', input_shape=(time_step, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=10, batch_size=16, verbose=1)

predicted = model.predict(X)
predicted = scaler.inverse_transform(predicted)
real = scaler.inverse_transform(y.reshape(-1, 1))

plt.figure(figsize=(12, 6))
plt.plot(real, label='Actual Stock Prices', color='blue')
plt.plot(predicted, label='Predicted Stock Prices', color='red', linestyle='--')
plt.title('Stock Price Prediction using RNN')
plt.xlabel('Time Step')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
